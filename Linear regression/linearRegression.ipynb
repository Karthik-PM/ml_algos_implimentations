cost function

```python
def costfunction(m,c,points):
    x = data['x']
    y = data['y']
    total_error = 0
    for i in range(0,len(points)):
        total_error += ((y.iloc[i] - (m*x.iloc[i] + c))**2)
    average_error = total_error/len(points)
    return average_error
```

minimizing cost function

```python
def gradiendDecent(m,c,points,L):
    gradiendDecent_m = 0
    gradiendDecent_n = 0

    n = len(points)
    
    for i in range(0,len(points)):
        x = data['x'].iloc[i]
        y = data['y'].iloc[i]
        gradiendDecent_m += -(2/n)* x *(y - (m*x + c))
        gradiendDecent_n += -(2/n)*(y - (m*x + c))
    m = m -  gradiendDecent_m*L 
    c = c - gradiendDecent_n*L
    return m,c
```

getting the optimized linear equation

```python
testcases = 300
m = 1
c = 1
L = 0.0001
for i in range(0,testcases):
    m,c = gradiendDecent(m,c,data,L)    
    test_result = []
    plt.scatter(data['x'],data['y'])
    plt.plot(list(range(5,77)),[m*i+c for i in range(5,77)],color = 'red')
    if( i % 50 == 0):
        print("y=",m,"x+",c)
        plt.show()


print(m,c)

#print(trained_result)
    
```
